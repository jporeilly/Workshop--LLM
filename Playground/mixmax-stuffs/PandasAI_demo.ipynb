Now we import the dependencies:

```python
import pandas as pd
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI
from pandasai.llm.open_assistant import OpenAssistant
```

We create a dataframe using pandas:

```python
df = pd.DataFrame({
    "country": ["United States", "United Kingdom", "France", "Germany", "Italy", "Spain", "Canada", "Australia", "Japan", "China"],
    "gdp": [21400000, 2940000, 2830000, 3870000, 2160000, 1350000, 1780000, 1320000, 516000, 14000000],
    "happiness_index": [7.3, 7.2, 6.5, 7.0, 6.0, 6.3, 7.3, 7.3, 5.9, 5.0]
})
```

```python
df
```

We instanciate the llm . Remember to change the API key with your own api key.
- https://huggingface.co/settings/tokens
- https://platform.openai.com/account/api-keys

```python
OPENAI_API_KEY = "YOUR_OPENAI_API_KEY"
llm = OpenAI(api_token=OPENAI_API_KEY)

# OpenAssistant
HF_API_KEY="YOUR_HUGGINGFACE_API_KEY"
llm_oa = OpenAssistant(api_token=HF_API_KEY)
```

### Pandasai with Open Assistant

Then we instantiate PandasAI with the provided large language model and we run it, passing the dataframe and the prompt

```python
pandas_ai_oa = PandasAI(llm_oa)
pandas_ai_oa.run(df, prompt='What is the sum of the GDPs of the 2 unhappiest countries?')
```

## Pandasai with OpenAI

```python
pandas_ai = PandasAI(llm)
pandas_ai.run(df, prompt='What is the sum of the GDPs of the 2 unhappiest countries based on the df provided?')
```

You can also show a chart, for example:

```python
pandas_ai.run(df, "I am looking for the data points that has happiness_index greater than 7")
```

```python
pandas_ai.run(df, "I am looking for the data points that has happiness_index greater than 7 and country name starts with alphabet C")
```

## Charts with Pandasai

```python
pandas_ai.run(df, "Plot the histogram of countries showing for each the gpd, using different colors for each bar")
```

```python
pandas_ai.run(df, "Plot the piechart of countries showing for each the gpd, using different colors for each bar")
```

```python
pandas_ai.run(df, "Plot the horizontal barplot of countries showing for each the gpd, using different colors for each bar")
```
